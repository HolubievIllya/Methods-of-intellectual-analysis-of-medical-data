```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Створити власний DataFrame

```python
df= pd.read_excel("samples.xlsx")
first_df = df[["Sample", "SPR response, ang. mdeg", "Response to regeneration, ang. mdeg"]]
first_df.head()
```

Побудувати для нього графіки (якщо відкрити графік на повне вікно буде краще видно точки)

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for ind in df.index:
    ax.scatter(df['SPR response, ang. mdeg'][ind], df['Response to regeneration, ang. mdeg'][ind], label=df['Sample'][ind])
colormap = plt.cm.gist_ncar
colorst = [colormap(i) for i in np.linspace(0, 0.9,len(ax.collections))]
for t,j1 in enumerate(ax.collections):
    j1.set_color(colorst[t])

ax.legend(fontsize='small')
plt.show()
```

Створити за даними власного DataFrame зведену таблицю


```python
pt = pd.pivot_table(df, index="Sample", values="SPR response, ang. mdeg", aggfunc="std")
pt.plot(kind="bar")
```

```python
pt1 = pd.pivot_table(df, index="Sample", values=["SPR response, ang. mdeg", "Response to regeneration, ang. mdeg"], aggfunc="max")
pt1.plot(kind="bar")
```
